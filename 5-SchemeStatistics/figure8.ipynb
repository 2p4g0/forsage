```python
import matplotlib
matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt
import pickle
import numpy as np
import pandas as pd
from datetime import datetime
x,y,sorted_blocks_first_seen_x, sorted_blocks_first_seen_y  = pickle.load(open("vals_over_time.pickle", "rb"))
```

```python
fig, ax = plt.subplots()

# Plot linear sequence, and set tick labels to the same color
blue = '#3F85FF'
orange = '#EF7700'
ax.plot(x,y, color=blue, linewidth=2)
ax.tick_params(axis='y', labelcolor=blue)
ax.set_xlabel("Date")
ax.set_ylabel("Total ETH")

# Generate a new Axes instance, on the twin-X axes (same position)
ax2 = ax.twinx()

# Plot exponential sequence, set scale to logarithmic and change tick color
ax2.plot(sorted_blocks_first_seen_x, sorted_blocks_first_seen_y, color=orange, alpha=0.8, linewidth=2)
ax2.tick_params(axis='y', labelcolor=orange)
ax2.set_ylabel("Total Count of Users")

plt.vlines(x=datetime.strptime('2020-02-01', '%Y-%m-%d'),ymin=0, ymax=max(sorted_blocks_first_seen_y), colors='black', alpha=0.3, linestyle='--')
ax.margins(x=0)
ax2.margins(x=0)
ax.set_xlim([datetime(2017, 6, 1, 0, 0, 0), datetime(2021, 1, 20, 0, 0, 0)])

import matplotlib.dates as mdate 
locator = mdate.YearLocator(1,1,1)
plt.gca().xaxis.set_major_locator(locator)
#locator = mdate.MonthLocator()
#plt.gca().xaxis.set_major_locator(locator)
formatter = mdate.ConciseDateFormatter(locator)
# formatter.formats = ['%Y']  # ticks are mostly years
# plt.gca().xaxis.set_major_formatter(formatter)
# fmt_month = mdate.MonthLocator()
# plt.gca().xaxis.set_minor_locator(fmt_month)
formatter.formats = ['%Y']  # ticks are mostly years
plt.gca().xaxis.set_major_formatter(formatter)
# fmt_month = mdate.MonthLocator()#1,1,1)
# plt.gca().xaxis.set_minor_locator(fmt_month)
# ax.grid(axis='x')
# ax2.grid(axis='x')


size = 12
plt.rcParams.update({'font.size': size})#, 'axes.labelsize': size})
ax.xaxis.label.set_size(size)
ax.yaxis.label.set_size(size)
ax2.xaxis.label.set_size(size)
ax2.yaxis.label.set_size(size)
ax.tick_params(axis='x', labelsize=size)
ax.tick_params(axis='y', labelsize=size)
ax2.tick_params(axis='x', labelsize=size)
ax2.tick_params(axis='y', labelsize=size)
#ax.set_yticklabels(['a','b','c','d', 'e','f','g'])
# labels = [v.get_text() for v in list(ax.get_yticklabels())]
labels=['−0.5M', '0', '0.5M', '1.0M', '1.5M', '2.0M', '2.5M', '3.0M']
ax.set_yticklabels(labels)
labels2=['−0.2', '0', '200k', '400k', '600k', '800k', '1M', '1.2M']
ax2.set_yticklabels(labels2)


# for v in labels:
#     v.set_text(v.get_text()+'M')
# labels2 = [str(item.get_text())+'M' for item in ax2.get_yticklabels()]
plt.show()
plt.tight_layout()
plt.savefig('/Users/Haaroony/Downloads/total_eth_recv_users_mid_2017_on_M.pdf')
```
